```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 問題1
# 平均
mean = np.array([-3, 0])
# 共分散行列
cov = np.array([[1.0, 0.8], [0.8, 1.0]])
n_of_seed = 500
np.random.seed(0)
seed_data_0 = np.random.multivariate_normal(mean, cov, n_of_seed)
print("平均: {}, 共分散行列: {}で生成した乱数のshape\n {}".format(mean, cov, np.shape(seed_data_0)))
```

```python
# 問題2
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy")
plt.scatter(seed_data_0[:,0], seed_data_0[:,1])
plt.show()
```

```python
# 問題3

# x1のグラフ
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("x1 dummy")
plt.xlim(-6, 3)
plt.hist(seed_data_0[:,0], bins=50)
plt.show()
# x2のグラフ
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("x2 dummy")
plt.xlim(-6, 3)
plt.hist(seed_data_0[:,1], bins=50)
plt.show()
```

```python
# 問題4
# 平均
mean_1 = np.array([0, -3])
# 共分散行列
cov_1 = np.array([[1.0, 0.8], [0.8, 1.0]])
n_of_seed = 500
np.random.seed(0)
seed_data_1 = np.random.multivariate_normal(mean_1, cov_1, n_of_seed)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy")
plt.scatter(seed_data_0[:,0], seed_data_0[:,1], label=0)
plt.scatter(seed_data_1[:,0], seed_data_1[:,1], label=1)
plt.legend()
plt.show()
```

```python
# 問題5

seed_data = np.vstack((seed_data_0, seed_data_1))
np.shape(seed_data)
```

```python
# 問題6

seed_data_0_with_label = np.column_stack((seed_data_0, np.full(len(seed_data_0), 0)))
seed_data_1_with_label = np.column_stack((seed_data_1, np.full(len(seed_data_1), 1)))
seed_data_with_label = np.vstack((seed_data_0_with_label, seed_data_1_with_label))
print(np.shape(seed_data_with_label))

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy")
plt.scatter(seed_data_with_label[:,0], seed_data_with_label[:,1])
plt.show()
```
