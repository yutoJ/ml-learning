```python
# 問題5
import matplotlib.pyplot as plt
%matplotlib inline

THICKNESS = 0.00008
number_of_folds = 43
folded_thickness_list = [THICKNESS]

for _ in range(number_of_folds):
    folded_thickness_list.append(folded_thickness_list[-1] * 2)

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list)
plt.show()
```

```python
# 問題6 1つ目
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folded_thickness_list, color='green')
plt.show()
```

```python
# 問題6 2つ目
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(folded_thickness_list)
plt.show()
```

```python
# 問題6 3つ目
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(direction='out', length=6, width=2, colors='b', grid_color='r', grid_alpha=0.5)
plt.plot(folded_thickness_list, color='green')
plt.show()
```
