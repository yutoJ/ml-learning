```python
# 倍化させる対象の体積が指定した体積を超えるまでの時間と超えるまでの増加の軌跡を計算
def compute_when_area_is_covered_manju_covered_with_doubled_object(doubled_object_volume, area_volume):
    #  初期値
    elapsed_time = 0
    total_doubled_object_volume = doubled_object_volume
    total_doubled_object_volume_list = [doubled_object_volume]
    
    # area_volumeを超えるまで倍化させていく
    while total_doubled_object_volume < area_volume:
        elapsed_time += 5
        total_doubled_object_volume *= 2
        total_doubled_object_volume_list.append(total_doubled_object_volume)
    
    return elapsed_time, total_doubled_object_volume_list
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import japanize_matplotlib

%matplotlib inline

# 太陽系の体積(m3)
area_volume = 1.5957856e+49
#　栗まんじゅうの体積(m3)
kuri_manju_volume = 10 ** -4

elapsed_time, total_kuri_manju_volume_list = compute_when_area_is_covered_manju_covered_with_doubled_object(kuri_manju_volume, area_volume)

# 結果
print("くりまんじゅう({}m^3)は、{}分後、{}m^3を超える".format(kuri_manju_volume, elapsed_time, area_volume))
# グラフ
plt.title("栗まんじゅうの増加傾向")
plt.xlabel("時間(分)")
plt.ylabel("体積(m3)")
plt.plot(list(range(0, elapsed_time + 5, 5)) , total_kuri_manju_volume_list, label="くりまんじゅうの体積")
plt.axhline(y=area_volume, color='r', linestyle=':', label='太陽系の体積')
plt.legend()
plt.show()
```

```python
# 東京ドームの体積(m3)
area_volume = 124 * 10000
#　ボールの体積(m3) ※20cm^3をm^3に変換
ball_volume = 20 / (100 * 100 * 100)

elapsed_time, total_volume_list = compute_when_area_is_covered_manju_covered_with_doubled_object(ball_volume, area_volume)
# 結果
print("ボール({}m^3)は、{}分後、{}m^3を超える".format(ball_volume, elapsed_time, area_volume))
# グラフ
plt.title("ボールの増加傾向")
plt.xlabel("時間(分)")
plt.ylabel("体積(m3)")
plt.plot(list(range(0, elapsed_time + 5, 5)) , total_volume_list, label="ボールの体積")
plt.axhline(y=area_volume, color='r', linestyle=':', label='東京ドームの体積')
plt.legend()
plt.show()
```
