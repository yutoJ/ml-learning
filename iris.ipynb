```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
# 問題1
X = pd.DataFrame(data.data, columns=["sepal_length", "sepal_width", "petal_length", "petal_width"])
print(X.describe())
print(X.head())
y = pd.DataFrame(data.target, columns=["Species"])
print(y.describe())
print(y.head())
```

```python
# 問題2
df = pd.concat([X, y], axis=1)
df.head()
```

```python
# 問題3 サンプル、データタイプ
print(df.head(4))
print(df.dtypes)
```

```python
# 問題3 ラベルごとのサンプル数の合計
print(df.groupby("Species").count())
print(df["Species"].value_counts())
```

```python
# 問題3 欠損値

# 全てfalseなので欠損なし
print(df.isnull().any())
# 全ての項目のnullが0なので欠損なし
print(df.isnull().sum())
```

```python
# 問題3 特徴量の平均値と標準偏差、四分位数
df.describe()
```

```python
# 問題4
print("irisの種類: {}".format(data.target_names))
print("データセットの説明: {}".format(data.DESCR))
print("項目の説明: {}".format(data.feature_names))
```

```python
# 問題5 sepal_widthの列を2つの異なる方法で抽出する
# 方法1
print(df["sepal_width"])
# 方法2
print(df.loc[:, "sepal_width"])
# 方法3
print(df.iloc[:, 1])
```

```python
# 問題5 50番目から99番目までのデータを抽出する
# 方法1
print(df[50:100])
# 方法2
print(df.loc[50:99])
# 方法3
print(df.iloc[50:100])
```

```python
# 問題5 
# 方法1
print(df.query("petal_width == 0.2"))
# 方法2
print(df[df["petal_width"] == 0.2])
```

```python
# 問題5 .locや.ilocの違い
print(".locはcolumn名を使って列を指定、.ilocはcolumnのindexを使って列を指定できる")
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
```

```python
# 問題6　円グラフ
plt.pie(df["Species"].value_counts(), labels=data.target_names)
plt.show()
```

```python
# 問題6　箱ひげ図1 (全体)
plt.boxplot(df.drop("Species", axis=1), labels=data.feature_names)
plt.show()
```

```python
# 問題6　箱ひげ図2 (個別)
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))
for i, feature in enumerate(data.feature_names):
    axs[i].boxplot(df.iloc[:,i], labels=[feature])
plt.show()
```

```python
# 問題6 箱ひげ図 (個別) seaborn
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))
for i, feature in enumerate(data.feature_names):
    sns.boxplot(x=df.iloc[:,i], ax=axs[i])
```

```python
# 問題6 バイオリン1 (全体)
plt.violinplot(df.drop("Species", axis=1))
plt.show()
```

```python
# 問題6 バイオリン2 (個別)
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))
for i, feature in enumerate(data.feature_names):
    axs[i].violinplot(df.iloc[:,i])
    axs[i].set_title(feature)
plt.show()
```

```python
# 問題6 バイオリン2 (個別) seaborn
fig, axs = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))
for i, feature in enumerate(data.feature_names):
    sns.violinplot(x=df.iloc[:,i], ax=axs[i])
    sns.stripplot(x=df.iloc[:,i], color="r", ax=axs[i])
```

```python
# 問題7 散布図 散布図行列 matplotlab
pd.plotting.scatter_matrix(df.drop("Species", axis=1), c=df["Species"])
plt.show()
```

```python
# 問題7 散布図 散布図行列 seaborn
sns.pairplot(df, hue="Species")
```
