```python
# 問題2 (関数)
def compute_sorori_shinzaemon(n_days=100):
    """曽呂利新左衛門の米の逸話における、日にちごとの米の粒の数を計算する関数。

    Parameteres
    -----------
    n_days: int
        米を貰う日数 (default : 100)

    Returns
    -------
    list_n_grains : list
        ある日に貰う米の数のリスト (リストの長さはn_daysに等しい)
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト (リストの長さはn_daysに等しい)
    """
    list_n_grains = []
    list_total_grains = []

    for n_day in range(1, n_days + 1):
        if n_day == 1:
            list_n_grains.append(1)
            list_total_grains.append(1)
        else:
            #  最後にもらった数の2倍の数をリストに追加
            list_n_grains.append(list_n_grains[-1] * 2)
            #  前日までの総和とその日にもらった数の和をリストに追加
            list_total_grains.append(list_total_grains[-1] + list_n_grains[-1])
        

    return list_n_grains, list_total_grains
```

```python
# 関数が正しいか確認
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=10)
print(list_n_grains)
print(list_total_grains)
```

```python
# 問題3
def compute_num_of_days_people_can_live(n_of_rice_grace, n_of_people):
    # お茶碗3倍分の米粒の数
    n_of_rice_grace_one_person_can_live = 3250 * 3
    
    # 米粒の数を人数と1日に必要な米粒の量で割った数の整数値がその日まで均等に分配して足りる日数
    return n_of_rice_grace // (n_of_people * n_of_rice_grace_one_person_can_live)

total_n_of_rice_grace = 100000
n_of_people = 3
print("米{}粒で{}人が{}日間、生活できる".format(total_n_of_rice_grace, n_of_people, compute_num_of_days_people_can_live(total_n_of_rice_grace, n_of_people)))
```
