```python
# 問題3

# 勾配　= yの変化量 / xの変化量
def  compute_slopes(x, y):
    x = np.arange(-50, 50 + 0.1, 0.1)
    return (y[1:] - y[:-1]) / (x[1:] - x[:-1])

slopes =  compute_slopes(x, y)
# print(slopes)
print(x[495:505])
print(slopes)
```

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

# 問題4

def display_formula_and_gradient_graph(x, y, gradient, formula):
    plt.title("1次関数")
    plt.plot(x, y, label=formula)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

    plt.title("勾配")
    plt.plot(x[1:], gradient)
    plt.xlabel('x')
    plt.ylabel('gradient')
    plt.show()
    
display_formula_and_gradient_graph(x, y, slopes,'y = (1 / 2) * x + 1')
```

```python
# 問題5 関数
def compute_gradient(formula_func, x_range=(-50, 50.1, 0.1)):
    """
    変化量を使い勾配を計算する。

    Parameters
    ----------------
    function : function
      求めたい関数、xのndarrayを入れるとyのndarrayを返すもの。
    x_range : tuple
      np.arange()の引数と同じ方法で範囲を指定する。

    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      xとyを結合したもの。nはx_rangeに依存する。
    gradient : ndarray, shape(n-1,)
      関数の勾配。差をとるため要素は1つ減る
    """
    x = np.arange(*x_range)
    y = formula_func(x)
    array_xy = np.column_stack([x, y])
    gradient = (y[1:] - y[:-1]) / (x[1:] - x[:-1])
    return array_xy, gradient
```

```python
# 問題5 y = (1/2) * x + 1
def linear_func(array_x):
    array_y = (1/2) * array_x + 1
    return array_y

array_xy, gradient = compute_gradient(linear_func)
display_formula_and_gradient_graph(array_xy[:,0], array_xy[:,1], gradient,'y = (1 / 2) * x + 1')
```

```python
# 問題5 y = x ** 2
def quadratic_func(array_x):
    array_y = array_x ** 2
    return array_y

array_xy, gradient = compute_gradient(quadratic_func)
display_formula_and_gradient_graph(array_xy[:,0], array_xy[:,1], gradient,'x ** 2')
```

```python
# 問題5 y = sin(x ** 1/2)
def sin_func(array_x):
    array_y = np.sin(array_x ** (1/2))
    return array_y

array_xy, gradient = compute_gradient(sin_func, x_range=(0, 50.1, 0.1))
display_formula_and_gradient_graph(array_xy[:,0], array_xy[:,1], gradient,'y = sin(x ** 1/2)')
```

```python
def find_y_min_argmin_and_gradients(formula_func, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = formula_func(x)
    return (x[np.argmin(y)], np.min(y), *compute_gradients_around_center(x, y, np.argmin(y)))

def compute_gradients_around_center(x, y, center_x_index):
    one_before_gradient = None
    one_after_gradient = None
    if center_x_index - 1 >= 0:
        one_before_gradient = (y[center_x_index] - y[center_x_index - 1]) / (x[center_x_index] - x[center_x_index - 1])
    if center_x_index + 1 <= len(x):
        one_after_gradient = (y[center_x_index + 1] - y[center_x_index]) / (x[center_x_index + 1] - x[center_x_index])
    return (one_before_gradient, one_after_gradient)
```

```python
# 問題6 y = (1/2) * x + 1
print("xが{}の時, yの最小値{}, 勾配は一つ前: {}, 一つ後: {}".format(*find_y_min_argmin_and_gradients(linear_func)))
```
